```python
# Q1. Key Statistical Definitions
## a) Population and Sample
# Population (all employees)
population = [23, 25, 28, 30, 32, 35, 38, 40, 42, 45]

# Sample (a few employees selected)
sample = [25, 30, 35, 40]

# Calculate mean age for population and sample
import numpy as np
pop_mean = np.mean(population)
sample_mean = np.mean(sample)

print("Population Mean:", pop_mean)
print("Sample Mean:", sample_mean)
```

```python
# b) Descriptive vs Inferential Statistics
# Descriptive: summarize the data (mean, median, std)
data = [12, 14, 18, 19, 21, 25, 30]

print("Mean:", np.mean(data))
print("Median:", np.median(data))
print("Standard Deviation:", np.std(data))

# Inferential: making predictions using t-test (example)
from scipy import stats
sample1 = [12, 14, 18, 19, 21]
sample2 = [20, 22, 25, 28, 30]

t_stat, p_val = stats.ttest_ind(sample1, sample2)
print("\nT-test result:")
print("t-statistic:", t_stat)
print("p-value:", p_val)
```

```python
# c) Parameter and Statistic
# Let's assume population (parameter known)
population = [50, 52, 54, 56, 58, 60]

# Sample drawn from population
sample = [52, 56, 58]

# Parameter (population mean)
parameter_mu = np.mean(population)

# Statistic (sample mean)
statistic_xbar = np.mean(sample)

print("Population Mean (Parameter):", parameter_mu)
print("Sample Mean (Statistic):", statistic_xbar)
```

```python
### Q5. Variance and Standard Deviation


# a) Definitions
# Variance (σ² or s²)
import numpy as np

# Dataset
data = [2, 4, 6]

# Variance
variance = np.var(data)  # population variance
sample_variance = np.var(data, ddof=1)  # sample variance (divided by n-1)

# Standard Deviation (σ or s)
std_dev = np.sqrt(variance)
sample_std_dev = np.sqrt(sample_variance)

print("Data:", data)
print("Population Variance:", variance)
print("Sample Variance:", sample_variance)
print("Population Standard Deviation:", std_dev)
print("Sample Standard Deviation:", sample_std_dev)
```

```python
# Step 1: Mean
mean = sum(data) / n

# Step 2: Sum of squared deviations
sq_diff = [(x - mean) ** 2 for x in data]
sum_sq_diff = sum(sq_diff)

# Step 3: Sample variance and standard deviation
sample_variance = sum_sq_diff / (n - 1)
sample_std = sample_variance ** 0.5

#Step 4: Population variance and standard deviation
population_variance = sum_sq_diff / n
population_std = population_variance ** 0.5
```

```python
print(f"Mean = {mean:.2f}")
print(f"Sample Variance = {sample_variance:.2f}")
print(f"Sample Std Deviation = {sample_std:.2f}")
print(f"Population Variance = {population_variance:.2f}")
print(f"Population Std Deviation = {population_std:.2f}")
```

```python
# Q8. Five-number Summary and Boxplot Concept
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# Dataset
data = [10, 12, 14, 16, 20, 25, 40]

# Convert to a NumPy array
data = np.array(data)

# Five-number summary
minimum = np.min(data)
q1 = np.percentile(data, 25)
median = np.median(data)
q3 = np.percentile(data, 75)
maximum = np.max(data)

# Interquartile Range (IQR)
iqr = q3 - q1

# Outlier bounds
lower_bound = q1 - 1.5 * iqr
upper_bound = q3 + 1.5 * iqr

# Display results
print("📊 Five-Number Summary:")
print(f"Minimum: {minimum}")
print(f"Q1 (25th percentile): {q1}")
print(f"Median (Q2): {median}")
print(f"Q3 (75th percentile): {q3}")
print(f"Maximum: {maximum}")
print(f"IQR: {iqr}")

print("\n📈 Outlier Detection:")
print(f"Lower Bound: {lower_bound}")
print(f"Upper Bound: {upper_bound}")

# Detect outliers
outliers = data[(data < lower_bound) | (data > upper_bound)]
if len(outliers) > 0:
    print(f"Outliers: {outliers}")
else:
    print("No outliers found.")

# Boxplot visualization
plt.figure(figsize=(6,4))
plt.boxplot(data, vert=False, patch_artist=True, boxprops=dict(facecolor='lightblue'))
plt.title("Boxplot - Five Number Summary")
plt.xlabel("Values")
plt.grid(True, linestyle="--", alpha=0.6)
plt.show()
```

```python
# Q10. Hypothesis Testing – One Sample Z-Test

import math

# Given data
mu_0 = 30000      # population mean
x_bar = 31000     # sample mean
sigma = 4900      # std dev
n = 49            # sample size
alpha = 0.05

# Step 1: Compute Z-score
Z = (x_bar - mu_0) / (sigma / math.sqrt(n))

# Step 2: Critical value for one-tailed test
Z_critical = 1.645

# Step 3: Decision
if Z > Z_critical:
    result = "Reject H0: Average salary has increased."
else:
    result = "Fail to reject H0: Not enough evidence to say salary increased."

print(f"Z-score: {Z:.2f}")
print(f"Critical Z (α=0.05): {Z_critical}")
print("Decision:", result)
```
